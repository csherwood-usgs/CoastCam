```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("D:/crs/proj/2019_CACO_CoastCam/decision_matrix_input.csv")
df
```

```python
value = df['Value'].values
valerr = df['Uncertainty'].values
diff = 0.25*df['Science'].values+0.25*df['Code'].values+0.25*df['Data'].values+0.25*df['Display'].values
differr = np.sqrt( 0.25*df['Science'].values**2 +0.25*df['Code'].values**2 +0.25*df['Data'].values**2 +0.25*df['Display'].values**2 )
label = df['Short name']
```

```python
plt.errorbar(diff, value, valerr/4, differr/4, 'ok', ecolor='dimgray', alpha = .4)
plt.xlim([0, 5])
plt.ylim([0, 5])
plt.yticks([])
plt.xticks([])
plt.axis('equal')
for i, (x,y) in enumerate(zip(diff,value)):
    plt.text(x+.1,y+.1,label[i], rotation=30)
plt.xlabel('Easy                           Hard')
plt.ylabel('Low benefit            High Benefit')
```
