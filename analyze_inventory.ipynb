```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('caco_rectify_stats.csv', header=0, delimiter=',',parse_dates={'datetime': [1]})
df
```

```python
df.describe()
```

```python
plt.plot(df['datetime'],df['s0'],'.')
```

```python
plt.plot(df['s0'],df['blur0'],'.')
plt.xlabel('Sharpness')
plt.ylabel('Blur')
```

```python
plt.plot(df['avg0'],df['s0'],'.')
plt.xlabel('Brightness')
plt.ylabel('Sharpness')
```

```python
plt.plot(df['avg0'],df['c0'],'.')
plt.xlabel('Brightness')
plt.ylabel('Contrast')
```

```python
plt.plot(df['s0'],df['s1'],'.')
plt.ylabel('Sharpness, c2')
plt.xlabel('Sharpness, c1')
```

```python
plt.plot(df['s0'],df['c0'],'.')
plt.ylabel('Contrast')
plt.xlabel('Sharpness')
```

```python
# what is the distribution of brightness difference?
diff = df['avg0'].values-df['avg1'].values
plt.hist(diff)
df['adiff'] = df['avg0']-df['avg1']
```

```python
# get a subsample with decent sharpness
dfs = df[ df['adiff']>7.]
dfs = dfs[ dfs['s0']>2.].sample(n=10)
dfs.to_pickle('tenfiles.pkl')
```
